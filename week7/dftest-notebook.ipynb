```python
# Dependancies 
import pandas as pd 
import numpy as np 
```

```python
weather_df = pd.DataFrame(np.random.rand(10,2)*5,
                          index=pd.date_range(start="2021-01-01", periods=10),
                          columns=["Tokyo", "Beijing"])
```

```python
# function for determining the ranges for weather conditions
def rain_condition(v):
    if v < 1.75:
        return "Dry"
    elif v < 2.75:
        return "Rain"
    return "Heavy Rain"

# function for making the output of the data frame formatted 
def make_pretty(styler):
    styler.set_caption("Weather Conditions")
    # apply the rain_conditions to format the data 
    styler.format(rain_condition)
    # passing %A to strftime gives the full name of the given day
    styler.format_index(lambda v: v.strftime("%A"))
    # setting background colors based on numeric values 
    styler.background_gradient(axis=None, vmin=1, vmax=5, cmap="YlGnBu")
    return styler
```

```python
print(weather_df)

weather_df.loc["2021-01-04":"2021-01-08"].style.pipe(make_pretty)
```

```python
weather_df.loc["2021-01-04":"2021-01-20"].style.pipe(make_pretty)
```
